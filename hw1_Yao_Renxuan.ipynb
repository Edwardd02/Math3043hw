```python
try:
    import numpy as np
except ImportError as e:
    print("Error: ", e)
```

```python
try:
    import matplotlib.pyplot as plt
except ImportError as e:
    print("Error: ", e)
```

```python
try:
    import copy
except ImportError as e:
    print("Error: ", e)
```

```python
# Create an array of evenly spaced x values on the interval [0, 2*pi]
x = np.linspace(0, 2*np.pi) # 50 elements in default
# Create an array of y1 values for the function y = sin(x)
y1 = np.sin(x)
# Create a plot of our data
plt.plot(x,y1)
# Show the plot
plt.show()
```

```python
# Store the values of y1 in y2
y2 = y1

# Multiply y1 by 2
y1 *= 2

# Plot the y2 data
plt.plot(x,y2)

# Show the plot
plt.show()
```

```python
# Reset the y1 values
y1 = np.sin(x)

# Point y2 to the same memory location as y1
y2 = y1

# Copy the y1 values into the y3 array
y3 = copy.deepcopy(y1)

# Multiply y1 by 2
y1 *= 2

# Plot the y2 and y3 data
plt.plot(x,y2)
plt.plot(x,y3)

# Create a legend
plt.legend(['y2', 'y3'])

# Show the plot
plt.show()
```

```python
# Load in the random module
from numpy import random

# Set the seed for random number generation, to ensure we get the same result each time
random.seed(99)

# Create a 15x10 array of random integers selected from 0 to 100
A = random.randint(100, size=(15,10))

# Print the array we have created
print(A)
```

```python
# Read the file into a new integer array B
B = np.fromfile('newArray.bin', dtype='int')

# Print array B
print(B)
```

```python
# Reshape vector B into array B
B = np.reshape(B, (15,10))

# Print array B
print(B)
```

```python
print(0.1 + 0.1 + 0.1 == 0.3)
```

<span style="font-size: 15px; color: red;">The result is False, it's about how computer represents float numbers. Since computer system uses binary, the float number can only be represent as $1/2 * b_1 + 1/4 * b_2 + 1/8 * b_3...$, so that there will be a tiny error for a huge part of float numbers. In such a case, we may try to compare the abs(diffrence) of the two float numbers with a very small number, which is shown in next cell.</span>


```python
print(abs((0.1 + 0.1 + 0.1 - 0.3)) < 1e-9)
```
