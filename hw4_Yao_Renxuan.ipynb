Solving an  n times n  system using Gaussian elimination requires approximately $ \dfrac{2n^3}{3}$multiplication and division operations.

## Proof Overview

Gaussian elimination involves:

1. **Forward Elimination**: Transform the matrix to upper triangular form.
2. **Back Substitution**: Solve for the variables.
5. **Row Pivoting**: This explains the case when it's used with scaled row pivoting

### 1. Forward Elimination

For each pivot row \( i = 1 \) to \( n - 1 \):

#### a. Compute Multipliers

For rows \( j = i + 1 \) to \( n \):

$$
m_{ji} = \dfrac{a_{ji}}{a_{ii}} \quad \text{(1 division per row)}
$$

- **Total divisions**:
  $$
  D_{\text{elim}} = \sum_{i=1}^{n-1} (n - i) = \dfrac{n(n - 1)}{2}
  $$

#### b. Update Rows

For columns \( k = i + 1 \) to \( n \):

$$
a_{jk} \leftarrow a_{jk} - m_{ji} \cdot a_{ik} \quad \text{(1 multiplication per element)}
$$

- **Total multiplications**:
  $$
  M_{\text{elim}} = \sum_{i=1}^{n-1} (n - i)(n - i) = \sum_{k=1}^{n-1} k^2 = \dfrac{n(n - 1)(2n - 1)}{6}
  $$

### 2. Back Substitution

For \( i = n \) down to \( 1 \):

#### a. Compute Sum

$$
s_i = \sum_{k=i+1}^{n} a_{ik} x_k \quad \text{(1 multiplication per term)}
$$

- **Total multiplications**:
  $$
  M_{\text{back}} = \sum_{i=1}^{n} (n - i) = \dfrac{n(n - 1)}{2}
  $$

#### b. Solve for \( x_i \)

$$
x_i = \dfrac{b_i - s_i}{a_{ii}} \quad \text{(1 division per variable)}
$$

- **Total divisions**:
  $$
  D_{\text{back}} = n
  $$

### 3. Total Operation Count

#### Divisions

$$
D_{\text{total}} = D_{\text{elim}} + D_{\text{back}} = \dfrac{n(n - 1)}{2} + n = \dfrac{n^2 + n}{2}
$$

#### Multiplications

$$
\begin{align*}
M_{\text{total}} &= M_{\text{elim}} + M_{\text{back}} \\
&= \dfrac{n(n - 1)(2n - 1)}{6} + \dfrac{n(n - 1)}{2} \\
&= \dfrac{2n^3 - 3n^2 + n}{6}
\end{align*}
$$

For large \( n \), the leading term dominates:

$$
M_{\text{total}} \approx \dfrac{2n^3}{3}
$$

### 4. Conclusion

The total number of multiplication and division operations is approximately:

$$
\text{Total Operations} = D_{\text{total}} + M_{\text{total}} \approx \dfrac{n^2}{2} + \dfrac{2n^3}{3} \approx \dfrac{2n^3}{3}
$$

**Therefore, Gaussian elimination requires approximately  $ \dfrac{2n^3}{3} $ long operations.**
### 5. Row Pivoting
**When it is used with scaled row pivoting, it involves $\dfrac{n^3}{3} + (\dfrac{1}{2}+m)n^2$ long operations**
### Efficient Solution of Multiple Linear Systems

When solving $A x^{(i)} = b^{(i)}$ for $i = 1, 2, \dots, m$ (where $m$ represents multiple systems with the same matrix $A$ but different right-hand sides), Gaussian elimination can be optimized by factoring $A$ into $PA = LU$. This factorization allows for efficient "back solves" to compute the solutions $x^{(i)}$.

#### Complexity Reduction
Instead of solving each system separately, which would require $\mathcal{O} \left( \frac{1}{3} m n^3 \right)$ operations, the LU factorization reduces the operation count. The forward elimination (LU factorization) phase takes $\mathcal{O} \left( \frac{1}{3} n^3 \right)$ operations, and the $m$ back solves require $\mathcal{O} \left( \left( \frac{1}{2} + m \right) n^2 \right)$ long operations in total. Therefore, the overall complexity is:

$
\mathcal{O} \left( \frac{1}{3} n^3 + \left( \frac{1}{2} + m \right) n^2 \right)
$

This is significantly less than solving each system individually.

```python
#Computer Problems
#4.3.2
import numpy as np
def symmetric_gaussian_elimination(A):
    n = A.shape[0]
    U = np.copy(A)  # Work on a copy of A to avoid modifying the original matrix

    for k in range(n - 1):        
        # Since the matrix is symmetric, work on the upper triangle and transpose results to the lower triangle
        for i in range(k + 1, n):
            factor = U[i, k] / U[k, k]
            U[i, k+1:] -= factor * U[k, k+1:]
            U[i, k] = 0  # Set the lower part to zero to maintain upper triangular form
            U[k+1:, i] -= factor * U[k+1:, k]  # Adjust the symmetric position

    return U

def solve_upper_triangular(U, b):
    n = U.shape[0]
    x = np.zeros(n)
    for i in reversed(range(n)):
        if abs(U[i, i]) < 1e-12:
            raise ValueError("Matrix is singular or nearly singular at diagonal element {}".format(i))
        x[i] = (b[i] - np.dot(U[i, i+1:], x[i+1:])) / U[i, i]
    return x

# Example usage
A = np.array([
    [0.6428, 0.3475, -0.8468],
    [0.3475, 1.8423, 0.4759],
    [-0.8468, 0.4759, 1.2147]
], dtype=float)
b = np.array([0.4127, 1.7321, -0.8621], dtype=float)

try:
    U = symmetric_gaussian_elimination(A)
    x = solve_upper_triangular(U, b)
    print("The solution vector x is:")
    print(x)
except Exception as e:
    print("Error:", e)
```

```python
#Problem 4.3.3
import numpy as np

def gaussian_elimination_with_pivoting(A, b):
    n = len(A)
    # Augment the matrix A with the vector b
    Ab = np.hstack([A, b.reshape(-1, 1)])

    for k in range(n):
        # Scale factors for each row
        scales = np.max(np.abs(Ab[:, :-1]), axis=1)
        # Perform pivoting based on the maximum ratio of |A[i, k]| / scale[i]
        if scales[k] == 0:
            raise ValueError("Singular matrix detected")
        pivot = np.argmax(np.abs(Ab[k:, k]) / scales[k:]) + k
        # Swap the current row with the pivot row
        Ab[[k, pivot]] = Ab[[pivot, k]]
        
        # Gaussian elimination
        for i in range(k+1, n):
            factor = Ab[i, k] / Ab[k, k]
            Ab[i, k:] -= factor * Ab[k, k:]

    # Back substitution
    x = np.zeros(n)
    for i in reversed(range(n)):
        x[i] = (Ab[i, -1] - np.dot(Ab[i, i+1:n], x[i+1:])) / Ab[i, i]
        if np.isnan(x[i]) or np.isinf(x[i]):
            raise ValueError("Singular matrix detected during back substitution")

    return x

# Example usage
A = np.array([
    [0.2641, 0.1735, 0.8642],
    [0.9411, 0.0175, 0.1463],
    [-0.8641, -0.4243, 0.0711]
])
b = np.array([-0.7521, 0.6310, 0.2501])

try:
    x = gaussian_elimination_with_pivoting(A, b)
    print("The solution vector x is:")
    print(x)
except Exception as e:
    print("Error:", e)
```
